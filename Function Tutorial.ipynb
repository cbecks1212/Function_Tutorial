```python
import pandas as pd
import ffn as ffn
from pylab import mpl, plt
import numpy as np
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
def fetch_manipulate_data(stocks, dt):
    data = ffn.get(stocks, start=dt)
    for col in data.columns:
        new_col = f"{col}_ret"
        data[new_col] = data[col].pct_change()
    clean = data.filter(like='ret')
    final_df = clean.dropna()
    return final_df
```

```python
def plot_returns():
    data[data.columns].cumsum().apply(np.exp).plot(figsize=(10,6), title='Comparing Returns')
```

```python
raw_input = input('enter tickers separated by comma')
stock_list = raw_input.split(',')
data= fetch_manipulate_data(stock_list, '2015-01-01')
plot_returns()
```
